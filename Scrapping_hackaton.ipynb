## Scraping des réponses

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.get("https://www.prix-carburants.gouv.fr/rubrique/faq/")

# Attendre que la page se charge complètement
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'fr-accordion__btn')))

# Récupérer tous les boutons de questions
buttons = driver.find_elements(By.CLASS_NAME, 'fr-accordion__btn')

# Liste pour stocker les réponses
reponses = []

# Boucle pour cliquer sur chaque question et récupérer la réponse
for button in buttons:
    try:
        # Cliquer sur la question pour afficher la réponse
        button.click()

        # Attendre un délai plus long après le clic sur le bouton (1 secondes)
        time.sleep(1)

        # Récupérer le texte de la réponse
        response_element = driver.find_element(By.XPATH, f'//div[@id="{button.get_attribute("aria-controls")}"]/p')
        text = response_element.text
        
        # Ajouter le texte récupéré à la liste des réponses
        reponses.append(text)
        
    except Exception as e:
        # Si une exception se produit, passer à la question suivante
        continue

# Fermer le navigateur
driver.quit()

# Afficher la liste des réponses
print(reponses)
```

## Création du dataframe

```python
import pandas as pd

# Create a dictionary with the 'question' and 'text' variables
data = {'question': question_modifiee, 'réponse': reponses}

# Create a dataframe from the dictionary
df = pd.DataFrame(data)

df
```

```python
df['url'] = 'https://www.prix-carburants.gouv.fr/rubrique/faq/'
df
```

```python
df_carburant = df
df_carburant['url'] = 'https://www.prix-carburants.gouv.fr/rubrique/faq/'
df_carburant
```

## Scraping des réponses

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la page à scraper
url = "https://recherche.data.gouv.fr/fr/faq"

# Faire une requête GET à la page
response = requests.get(url)

# Vérifier si la requête a réussi
if response.status_code == 200:
    # Analyser le contenu HTML de la page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Trouver tous les éléments avec l'ID commençant par "accordion-"
    reponses = soup.find_all(id=lambda x: x and x.startswith("accordion-"))

    # Créer une liste pour stocker les réponses
    liste_reponses = []

    # Parcourir toutes les réponses trouvées
    for reponse in reponses:
        # Ajouter le texte de la réponse à la liste
        liste_reponses.append(reponse.text.strip())

liste_reponses
```

## Création du dataframe

```python
import pandas as pd

# Create a dictionary with the 'question' and 'réponse' variables
data = {'question': question, 'réponse': liste_reponses}

# Create a DataFrame from the dictionary
df_1 = pd.DataFrame(data)

df_1
```

```python
df_1['url'] = 'https://recherche.data.gouv.fr/fr/faq'
df_1
```

```python
df_datagouv = df_1
df_datagouv
```

## Scraping des questions et des réponses

```python
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)
driver.get('https://www.banque-france.fr/fr/aide-faq/Moyens%20de%20paiement')
lien= driver.find_elements(By.XPATH, '//a[@class="py-0 text-black nav-link align-items-center "]')
links = [link.get_attribute('href') for link in lien]
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)

questions_BDF = []
reponses_BDF = []
for lien in links:
    if isinstance(lien, str) and lien.strip():  
        driver.get(lien)
        time.sleep(2)
    else:
        print(f'No link found')
    try:
        questions = driver.find_elements(By.XPATH, '//h3[@class="h5 d-flex align-items-center flex-grow-1 mb-0 py-5 me-1"]')
        answers = driver.find_elements(By.XPATH, '//div[@class="accordion-body"]/p')
        if questions:
            questions_BDF.append([elem.text for elem in questions if elem.text != ''])
        else:
            print(f'No questions in {lien}')
        if answers:
            reponses_BDF.append([elem.text for elem in answers if elem.text != ''])
        else:
            print(f'No answers in {lien}')
    except Exception as e:
        print(f"Error processing {lien}: {str(e)}")
```

```python
questions_BDF, len(questions_BDF)
```

```python
reponses_BDF, len(reponses_BDF)
```

```python

df_BDF = pd.DataFrame({'question': questions_BDF, 'réponse': reponses_BDF})
df_BDF
```

```python
df_BDF['url'] = 'https://www.banque-france.fr/fr/aide-faq/Moyens%20de%20paiement'
df_BDF
```

# Concours BCE

```python
import requests as req
```

## Scraping des questions et des réponses 

```python
req = req.get('https://www.concours-bce.com/la-foire-aux-questions')

html = req.text
soup= BeautifulSoup(html,'html.parser')
noms = soup.findAll('a',{'class':'trigger'})
questions_Bce = [noms.get_text() for noms in noms]
questions_Bce
rep = soup.findAll('div',{'class':'faq-content'})
réponses_bce = [rep.get_text() for rep in rep]
réponses_bce 
```

```python
questions_Bce, len(questions_Bce)
```

```python
reponse_BCE_modifiee = [réponses_bce[i][1:] for i in range(len(réponses_bce))]
reponse_BCE_modifiee, len(reponse_BCE_modifiee)
```

## Création du dataframe

```python
df_2 = pd.DataFrame({'question': questions_Bce, 'réponse': reponse_BCE_modifiee})
df_2
```

```python
df_2['url'] = 'https://www.concours-bce.com/la-foire-aux-questions'
df_2
```

```python
df_concours_BCE = df_2
df_concours_BCE
```

```python
# Liste des DataFrames à concaténer
dataframes = [df_carburant, df_datagouv, df_BDF, df_concours_BCE]

# Utilisez la fonction concat pour concaténer les DataFrames
df_final = pd.concat(dataframes, ignore_index=True)
df_final
```
